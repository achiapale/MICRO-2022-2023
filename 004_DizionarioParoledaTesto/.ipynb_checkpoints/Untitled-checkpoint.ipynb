```python
#CREARE UN DIZIONARIO DOVE LE CHIAVI SONO LE SINGOLE PAROLE
#E IL VALORE E' IL NUMERO DELLE VOLTE IN CUI SI RITROVA LA PAROLA ALL'INTERNO DEL TESTO
```

```python
from collections import Counter
```

```python
#apro il file e metto le righe in una lista righe[]
file = open("file.txt", "r")
righe = file.readlines()

#trasformo la lista in un testo
stringa = "".join(righe)

#attraverrso la funzione lower() trasformo il testo in tutte lettere minuscole
caratteriMinuscoli = stringa.lower()

#elimino la punteggiatura
listaPunteggiatura = [".", ",", ";", "!", "?", ":", "(", ")"]
eliminaPunteggiatura = [p for p in caratteriMinuscoli if p not in listaPunteggiatura]

#trasformo la lista in un testo
testSenzaPunteggiatura = " ".join(eliminaPunteggiatura)

#splitto il testo per dividere ogni singola parola
stringSplit = caratteriMinuscoli.split()

#attraverso la funzione counter, creo un dizionario dove salvo le parole e le volte in cui
#sono state ripetute nel testo in odine crescente
dizionario = Counter(stringSplit)

#stampo il dizionario
print(dizionario)

#chiudo il file
file.close()
```
