# FUNZIONI PRINCIPALI DI MATPLOTLIB

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dim = 200
immagine = np.random.uniform(0., 1., (dim, dim))
# creazione di un'immagine in bianco e nero
plt.imshow(immagine, cmap = "Greys")
```

```python
immagine = np.ones((dim, dim))
# fisso il valore minimo e massimo (1 = nero, 0 = bianco)
plt.imshow(immagine, cmap = "Greys", vmin = 0, vmax = 1)
```

```python
# punto bianco al centro
immagine = np.ones((dim, dim))
immagine[dim // 2, dim // 2] = 0
plt.imshow(immagine, cmap = "Greys", vmin = 0, vmax = 1)
```

```python
# righe (o colonne) tutte dello stesso colore
immagine = np.ones((dim, dim))
immagine[:, dim // 2] = 0
plt.imshow(immagine, cmap = "Greys", vmin = 0, vmax = 1)
```

```python
# banda colorata
immagine = np.ones((dim, dim))
immagine[:, 10:50] = 0
plt.imshow(immagine, cmap = "Greys", vmin = 0, vmax = 1)
```

```python
# quadrato
immagine = np.ones((dim, dim))
immagine[10:50, 10:50] = 0
plt.imshow(immagine, cmap = "Greys", vmin = 0, vmax = 1)
```
